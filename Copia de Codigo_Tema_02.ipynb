```
# Creando una variable string
Mensaje = "Hola"

# Imprimiendo la variable
print(Mensaje)

print("\t¿Como estas? \n \tEstoy bien", Mensaje)
# \n es un salto de linea
# \t es un tabulador
```

```
# Concatenacion de strings
M_1 = "Estoy"
M_2 = " dando"
M_3 = " una clase"
M_total = M_1 + M_2 + M_3

print(M_total)
print(M_1 + M_2 + M_3 + " de Python")

# No hagan esto
Numero_1 = "1"
Numero_2 = "2"
Numero_3 = Numero_1 + Numero_2
print(Numero_3)
```
